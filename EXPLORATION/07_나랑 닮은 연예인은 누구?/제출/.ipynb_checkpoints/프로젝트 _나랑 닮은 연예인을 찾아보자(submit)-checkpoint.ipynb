## 얼굴 임베딩 추출

```python
import face_recognition
import os

# 사진에서 얼굴 크롭하는 함수 만들기
def get_gropped_face(image_file):
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    a, b, c, d = face_locations[0]
    cropped_face = image[a:c,d:b,:]
    
    return cropped_face
```

```python
# 사진에서 얼굴 크롭하기
image_path = os.getenv('HOME')+'/aiffel/face_embedding/celebrity/민기.jpg'
cropped_face = get_gropped_face(image_path)

plt.imshow(cropped_face)
```

이렇게 사진에서 얼굴을 크롭해서 보여준다.

빗살무늬토기같이 고운 것이 아주 잘생겼다.

```python
import os

# 모아놓은 사진으로 리스트 만들기
dir_path = os.getenv('HOME')+'/aiffel/face_embedding/celebrity'
file_list = os.listdir(dir_path)

print(len(file_list))
print ("file_list: {}".format(file_list))
```

총 771명의 사진을 모았다.

```python
# 크롭된 얼굴 사진으로 임베딩 구하기
image_file = os.path.join(dir_path, '민기.jpg')
face = get_gropped_face(image_file)

# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)  
embedding, len(embedding[0])
```

내 얼굴은 이런 128차원의 숫자들로 이루어져 있다.

```python
# 파일의 리스트를 임베딩한 벡터와 함께 딕셔너리로 만들어주기
def get_face_embedding_dict(dir_path):
    err_cnt = 0
    file_list = os.listdir(dir_path)
    embedding_dict = {}
    cropped_dict = {}

    for file in file_list:
        img_path = os.path.join(dir_path, file)
        try:
            face = get_gropped_face(img_path)

            embedding = get_face_embedding(face)

            if len(embedding) > 0:
                embedding_dict[os.path.splitext(file)[0]] = embedding[0]
                cropped_dict[os.path.splitext(file)[0]] = face
        except:
            err_cnt += 1
            print(file)
    print(err_cnt)
    return embedding_dict, cropped_dict
```

```python
dir_path = os.getenv('HOME')+'/aiffel/face_embedding/celebrity'
embedding_dict, cropped_dict = get_face_embedding_dict(dir_path)
print(len(embedding_dict))
print(len(cropped_dict))
embedding_dict['공유']
```

## 얼굴 사진 비교하기

```python
def get_distance(name1, name2):
    return np.linalg.norm(embedding_dict[name1]-embedding_dict[name2], ord=2)

get_distance('공유', '민기')
```

나는 공유형님을 정말 좋아한다. (~*그래서 커피는 카누만 먹어요 형*~)

나와 공유형의 거리는 딱 이만큼...

```python
# name1과 name2의 거리를 비교하는 함수를 생성하되, name1은 미리 지정하고, name2는 호출시에 인자로 받도록 합니다.
def get_sort_key_func(name1):
    def get_distance_from_name1(name2):
        return get_distance(name1, name2)
    return get_distance_from_name1

sort_key_func = get_sort_key_func('공유')   
```

```python
sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))
```

```python
# 닮은 사람 출력해주는 함수 만들기
def get_nearest_face(name, top=5):
    sort_key_func = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))

    for i in range(top+1):
        if i == 0 :
            continue
        if sorted_faces[i]:
            print('순위 {} : 이름({}), 거리({})'.format(i, sorted_faces[i][0], sort_key_func(sorted_faces[i][0])))
    return sorted_faces
```

```python
# 가장 닮은 사람 출력하기
sorted_face = get_nearest_face('민기')
```

## 시각화

```python
# 순위에 맞는 이미지 출력
def get_nearest_face_images(sorted_faces, top=5):
    fig = plt.figure(figsize=(15, 5))
    fig.add_subplot(2, top, 1)
    plt.imshow(cropped_dict[sorted_faces[0][0]])
    for i in range(1, top+1):
        fig.add_subplot(2, top, i+5)
        plt.imshow(cropped_dict[sorted_faces[i][0]])
```

```python
get_nearest_face_images(sorted_face)
```
